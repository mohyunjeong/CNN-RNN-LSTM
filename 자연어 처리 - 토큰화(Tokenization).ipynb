```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from konlpy.tag import Okt
```

```python
example = "Family is not an important thing. It's everything."
stop_words = set(stopwords.words('english'))

word_tokens = word_tokenize(example)

result = []
for word in word_tokens:
    if word not in stop_words:
        result.append(word)

print('ë¶ˆìš©ì–´ ì œê±° ì „ : ', word_tokens)
print('ë¶ˆìš©ì–´ ì œê±° í›„ : ', result)
```

```python
# ë¬¸ìì—´ì´ ì£¼ì–´ì§„ ë¬¸ì ì§‘í•©(ì†Œë¬¸ì, a-z, ëŒ€ë¬¸ì A-Z, ìˆ«ì 0-9)ì˜ ë¬¸ìë§Œ í¬í•¨í•˜ëŠ”ì§€ í™•ì¸í•˜ëŠ” í•¨ìˆ˜

def check(string) :
    c = re.compile(r'[^a-zA-Z0-9.]')
    s = c.search(string)

    return not bool(s)

print(check('KimDoYoung031002'))
```

```python
# 1. ë¬¸ìì—´ì´ ì£¼ì–´ì§„ ë¬¸ì ì§‘í•©(ì†Œë¬¸ì aì™€ 0ê°œ ì´ìƒì˜ ì†Œë¬¸ì b)ì˜ ë¬¸ìë§Œ í¬í•¨í•˜ëŠ”ì§€ í™•ì¸í•˜ëŠ” í•¨ìˆ˜

def check(string) :
    patterns = 'ab*?'

    if re.search(patterns, string) :
        return 'True'
    else :
        return 'False'

print(check('abc'))
```

```python
# 2. ë¬¸ìì—´ì´ ì£¼ì–´ì§„ ë¬¸ì ì§‘í•©(ì†Œë¬¸ì aì™€ ë‘˜ í˜¹ì€ ì„¸ ê°œì˜ ì†Œë¬¸ì b)

def check(string):
    patterns = 'ab{2,3}'

    if re.search(patterns, string) :
        return 'True'
    else :
        return 'False'

print(check('abbc'))
```

```python
# 3. ë¬¸ìì—´ì´ ì£¼ì–´ì§„ ì§‘í•©('_'ë¡œ ì—°ê²°ëœ ì†Œë¬¸ì a-zì˜ ì§‘í•©)ë§Œ í¬í•¨í•˜ëŠ”ì§€ í™•ì¸í•˜ëŠ” í•¨ìˆ˜
# ì˜ˆì‹œ : aab_cbbc

def check(string) :
    patterns = '^[a-z]+_[a-z]+$'

    if re.search(patterns, string) :
        return 'True'
    else :
        return 'False'

print(check('a_b'))
```
